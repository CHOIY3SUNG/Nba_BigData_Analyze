```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file containing the statistics and outcomes
stats_and_outcomes = pd.read_csv("nba_stats.csv")

# Calculate the overall average statistics for winning games
overall_winning_stats = stats_and_outcomes[stats_and_outcomes["WL"] == 1].mean()
overall_winning_stats["FG_PCT"] = overall_winning_stats["FG_PCT"] * 100

# Select the statistics columns to plot
columns_to_plot = ["PTS", "AST", "REB", "TOV", "FG_PCT", "FG3A", "FGA", "PF", "BLK"]

# Plot the overall average statistics for winning games as a bar chart
plt.figure(figsize=(10, 6))
overall_winning_stats[columns_to_plot].plot(kind="bar")
plt.title("Overall Average Statistics for Winning Games")
plt.xlabel("Statistic")
plt.ylabel("Average")
plt.xticks(rotation=45)
plt.show()
print(overall_winning_stats)
```
