```python
import pandas as pd
import matplotlib.pyplot as plt

stats = pd.read_csv("all_stats.csv")

winning_stats = stats[stats["WL"] == 1].mean()
losing_stats = stats[stats["WL"] == 0].mean()

# Calculate the percentages for shooting statistics
winning_stats["FG_PCT"] *= 100
winning_stats["FG3_PCT"] *= 100
winning_stats["FT_PCT"] *= 100
losing_stats["FG_PCT"] *= 100
losing_stats["FG3_PCT"] *= 100
losing_stats["FT_PCT"] *= 100

# Create a DataFrame for the grouped bar graph
comparison_stats = pd.DataFrame({
    "Winning Games": winning_stats,
    "Losing Games": losing_stats
})

# Plot the grouped bar graph
plt.figure(figsize=(12, 6))
comparison_stats.plot(kind="bar")
plt.title("Comparison of Average Statistics: Winning Games vs. Losing Games")
plt.xlabel("Statistic")
plt.ylabel("Average")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

stats = pd.read_csv("all_stats.csv")

overall_winning_stats = stats[stats["WL"] == 1]
overall_winning_stats["FG_PCT"] = overall_winning_stats["FG_PCT"] * 100
overall_winning_stats["FG3_PCT"] = overall_winning_stats["FG3_PCT"] * 100
overall_winning_stats["FT_PCT"] = overall_winning_stats["FT_PCT"] * 100

columns_to_plot = [ "FGA", "FG_PCT", "FG3M", "FG3A", "FG3_PCT", "FTM", "FTA", "FT_PCT", "OREB", "DREB", "REB", "AST", "STL", "BLK", "TOV", "PF"]

# Calculate correlation coefficients between each statistic and winning
correlations = {}
for column in columns_to_plot:
    correlation = stats[column].corr(stats["WL"])
    correlations[column] = correlation

# Convert the correlations to a pandas Series
correlations = pd.Series(correlations)

# Sort the correlations in descending order
sorted_correlations = correlations.sort_values(ascending=False)

# Create a bar plot of the correlations
plt.figure(figsize=(10, 6))
sorted_correlations.plot(kind="bar")
plt.title("Correlation Coefficients between Statistics and Winning")
plt.xlabel("Statistic")
plt.ylabel("Correlation Coefficient")
plt.xticks(rotation=45)
plt.show()

# Print the correlation coefficients
print("Correlation Coefficients (Statistic - WL):")
print(sorted_correlations)
```

```python
import pandas as pd

stats = pd.read_csv("all_stats.csv")

overall_winning_stats = stats[stats["WL"] == 1].mean()
overall_losing_stats = stats[stats["WL"] == 0].mean()

overall_winning_stats["FG_PCT"] = overall_winning_stats["FG_PCT"] * 100
overall_winning_stats["FG3_PCT"] = overall_winning_stats["FG3_PCT"] * 100
overall_winning_stats["FT_PCT"] = overall_winning_stats["FT_PCT"] * 100

overall_losing_stats["FG_PCT"] = overall_losing_stats["FG_PCT"] * 100
overall_losing_stats["FG3_PCT"] = overall_losing_stats["FG3_PCT"] * 100
overall_losing_stats["FT_PCT"] = overall_losing_stats["FT_PCT"] * 100

columns_to_plot = ["FGM", "FGA", "FG_PCT", "FG3M", "FG3A", "FG3_PCT", "FTM", "FTA", "FT_PCT", "OREB", "DREB", "REB", "AST", "STL", "BLK", "TOV", "PF"]

# Calculate correlation coefficients between each statistic and winning
correlations = {}
for column in columns_to_plot:
    correlation_winning = stats[column].corr(stats["WL"])
    correlation_losing = stats[column].corr(1 - stats["WL"])
    correlations[column] = [correlation_winning, correlation_losing]

# Create a DataFrame for correlations
correlations_df = pd.DataFrame(correlations, index=["Winning", "Losing"])

# Sort the correlations in descending order
sorted_correlations = correlations_df.T.sort_values(by=["Winning"], ascending=False)

print("Statistics Correlation Coefficients:")
print(sorted_correlations)
```
